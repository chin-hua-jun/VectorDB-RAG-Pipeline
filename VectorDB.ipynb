```python
def inspect_chunks(chunks, n=5):
    """Print a sample of chunks for visual inspection in the notebook."""
    print(f"Total chunks: {len(chunks)}")
    print(f"Avg chunk size: {sum(len(c.page_content) for c in chunks) // len(chunks)} chars\n")
    for i, chunk in enumerate(chunks[:n]):
        print(f"--- Chunk {i + 1} | {chunk.metadata.get('source', 'unknown')} ---")
        print(chunk.page_content[:400])
        print()
```
